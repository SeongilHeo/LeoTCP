```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame(index=list(range(100,1001,100)))

result_dir=f"./26/result"

FD = list(range(0,12,2))
for freeze_duration in FD:
    freeze_duration  *= 0.1
    file_name=f"result_{freeze_duration:.1f}.csv"

    t_df = pd.read_csv(f"{result_dir}/{file_name}")

    names=['Algorithm Name','Bandwidth (Mbps)','Throughput (Mbps)','Loss Rate (%)']
    t_df.rename(columns=dict(zip(t_df.columns,names)),inplace=True)

    algo, bw, th, loss = t_df.columns
    dtype={names[0]: str, names[1]: float, names[2]: float, names[3]: float}
    t_df.astype(dtype)

    t_df.drop(algo,axis=1, inplace=True)
    t_df.set_index(bw, inplace=True)
    df[f"satcp_{freeze_duration:.1f}"]=t_df[th]

df.plot(xlabel=bw, ylabel=th,
         grid=True, 
         marker="o")
```
